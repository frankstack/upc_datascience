```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("student_survey.csv")
```

```python
df_randomsample = df.sample(n=192)
```

***

```python
#lateralidad poblacion
df.Handed.value_counts().to_frame("side").plot.pie(y="side")
```

```python
#lateralidad muestra
df_randomsample.Handed.value_counts().to_frame("side").plot.pie(y="side")
```

***

```python
#sexo poblacion
df.Sex.value_counts().to_frame("sex").plot.pie(y="sex")
```

```python
#sexo muestra
df_randomsample.Sex.value_counts().to_frame("sex").plot.pie(y="sex")
```
